```python
from google.colab import output
import time
user=True
turn=0
board=[
       ['_','_','_'],
       ['_','_','_'],
       ['_','_','_']
     ]
dummy_board=[
       ['1','2','3'],
       ['4','5','6'],
       ['7','8','9']
     ]
def printboard(board):
  for row in board:
    print('-------------------\n|',end="")
    for col in row:
      print(f'  {col}  |',end="")
    print()
def c_user(user):
  if user:return 'x'
  else: return 'O'

def check_input(user_input):
  if not user_input.isnumeric():
    print("not valid input\n")
    return False
  elif int(user_input)>=1 and int(user_input)<=9:
    return True
  else:
    print("not in bound\n")
    return False

def coordinates(user_input):
  row = int(user_input / 3)
  col = user_input
  if col > 2: col = int(col % 3)
  return (row,col)
def already_exist(co_ord):
  x=co_ord[0]
  y=co_ord[1]
  if board[x][y]!='_':
    return True
  else :return False

def update_board(user,board,co_ord):
  board[co_ord[0]][co_ord[1]]=user

def iswin(user,board):
  if rowcheck(user,board):return True
  elif colcheck(user,board): return True
  elif digcheck(user,board):return True
  else: return False
def rowcheck(user,board):
  for row in board:
    flag=True
    for col in row:
      if col!=user:
        flag=False
    if flag :return True
    else:return False
def colcheck(user,board):
  for col in range(3):
    flag=True
    for row in range(3):
      if board[row][col]!=user:
        flag=False
    if flag:return True
    else:return False
def digcheck(user,board):
  if board[0][0] == user and board[1][1] == user and board[2][2] == user: return True
  elif board[0][2] == user and board[1][1] == user and board[2][0] == user: return True
  else: return False
def reload_board(board):
  for row in range(3):
    for col in range(3):
      board[row][col]='_'
#printboard(dummy_board)
```

```python
def start_game(user=True,turn=0):
  while turn<=9:
    active_user=c_user(user)
    printboard(board)
    print(f'It is your turn: {active_user}\n')
    user_input=input("enter your place:1 to 9 .\n \n press q for Quit\n")
    if user_input=='q':break
    if not check_input(user_input):
      print("Try again:")
      continue
    user_input = int(user_input)-1
    co_ord=coordinates(user_input)
    if already_exist(co_ord):
      print("Already filled try other position\n")
      continue
    update_board(active_user,board,co_ord)
    if iswin(active_user,board):
      printboard(board)
      print(f'****Game won by {active_user}!******\n')
      print('Thank you!\n')
      break
    if turn==9:
      print("game is tie!!\n")
      break
    turn+=1
    user =not user
    

```

```python
print("Welcome to tic toe game")
while True:
  printboard(dummy_board)
  reload_board(board)
  print("\n \n above represents the number for each tic toe game \n Please remember it")
  start_game()
  quit= input("play again y/n?")
  if quit=='n':
    break
  #7
  reload_board(board)
```
